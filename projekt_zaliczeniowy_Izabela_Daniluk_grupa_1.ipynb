```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
#stopwords angielskie
stop = set(stopwords.words("english"))
tweets["text_without_stopwords"] = tweets["text"].apply(lambda x: " ".join([elem for elem in x.split() if elem not in (stop)]))
```

```python
#stopwords polskie
with open("stopwords.txt", "r") as f:
 new_stopwords = []
 for line in f.readlines():
  new_stopwords.append(line)

new_stopwords_set = set(new_stopwords)
polskie_stopwords = [elem.replace("\n", "") for elem in new_stopwords_set]
tweets["text_without_stopwords_polskie"] = tweets["text"].apply(lambda x: " ".join([elem for elem in x.split() if elem.lower() not in (polskie_stopwords)]))
```

```python
#3
nlp = spacy.load("pl_core_news_sm")
nlp.pipeline

tweets["text_soacy"] = tweets["text"].apply(nlp)

doc = tweets["text_soacy"]
doc1 = "".join(str(x) for x in doc)
text = nlp(doc1)
```

```python
persons = [ent.text for ent in text.ents if ent.label_ == "persName"]
tweets["persons"] = pd.Series(persons)
tweets["persons"]
```

```python
places = [ent.text for ent in text.ents if ent.label_ == "placeName"]
tweets["places"] = pd.Series(places)
tweets["places"]
```

```python
organizations = [ent.text for ent in text.ents if ent.label_ == "orgName"]
tweets["organizations"] = pd.Series(organizations)
tweets["organizations"]
```
